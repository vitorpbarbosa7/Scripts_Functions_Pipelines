```python
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('granulometrica.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df2 = pd.melt(df, id_vars="Tamanhos", var_name="Materiais", value_name="Diferencial")
```

```python
df2.sample(10)
```

```python
sea.scatterplot(df2.Tamanhos, df2.Diferencial, hue=df2.Materiais)
```

```python
f, ax = plt.subplots(figsize = (7,7))
ax.set(xscale = "log")
sea.lineplot(df2.Tamanhos, df2.Diferencial, hue=df2.Materiais)
```
